```python
import os
import json
from tabulate import tabulate
```

```python
root_folder = r"C:\Users\hp\Desktop\Module-3-Task-assigning\nn_results"  # <-- Change this to your actual path

# To store all results
results = []

# Loop through each item in the root folder
for project_folder in os.listdir(root_folder):
    project_path = os.path.join(root_folder, project_folder)
    if os.path.isdir(project_path) and project_folder.endswith("_features"):
        # Extract project name (e.g., AMBARI_5)
        project_name = "_".join(project_folder.split("_")[2:4])
        results_file = os.path.join(project_path, "results_nn.json")

        if os.path.isfile(results_file):
            with open(results_file, "r") as f:
                data = json.load(f)
                results.append([
                    project_name,
                    round(data.get("accuracy", 0), 4),
                    round(data.get("loss", 0), 4),
                    round(data.get("precision_macro", 0), 4),
                    round(data.get("recall_macro", 0), 4),
                    round(data.get("f1_score_macro", 0), 4),
                ])

# Print results in tabular format
headers = ["Project Name", "Accuracy", "Loss", "Precision", "Recall", "F1 Score"]
print(tabulate(results, headers=headers, tablefmt="grid"))
```
